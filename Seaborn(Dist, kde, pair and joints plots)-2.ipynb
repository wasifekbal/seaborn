```python
# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.
iris = sns.load_dataset('iris')
iris.head(1)
sns.pairplot(iris,hue='species')
```

Conclusions :-
1. From petal_length we can easily distinguish Setosa with Versicolor and virginica.
2. The size of the petal_width, petal_length and sepal_lenght of Satosa is largerst among these three species.
3. Virginica has the largest sepal_width among others.
4. Sepal_width of Versicolor and virginica is almost fully overlapped with each other, which means the probability of getting same Sepal_width for both is high.
5. Petal_width and Petal_length of all three species are linearly related with each other, with an order Setosa < Versicolor < virginica. 
6. Petal_length and Sepal_length of Versicolor and virginica are linearly related with each other, but Setosa is not.

```python
# Problem 3: Plot a joint plot showing relation between 'petal length' and 'petal width' with density estimates.
sns.jointplot(x='petal_length',y='petal_width',kind='kde',height=8,data=iris)
```
