```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('titanic')
```

```python
plt.style.use('fivethirtyeight')
data.head()
```

```python
# Problem 1 : Plot a Heatmap on the correlation between multiple columns
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data.corr(), annot=True, cbar=True, cmap="summer").set_ylim(8.0,0)
plt.title("Heatmap of the correlation between all the columns of Titanic dataset\n")
plt.show()
```

```python
# Problem 2 : Plot the distribution of age column using a Boxplot on the basis of sex column
```

```python
sns.catplot(x='sex',y='age',kind='box',height=6,aspect=1.5,data=data)
```

```python
# Problem 3 : Plot the distribution of age using a ViolinPlot on the basis of both Pclass and Sex column
```

```python
sns.catplot(y='age',x='pclass',hue='sex',kind='violin',aspect=1.6,height=6,data=data,split=True)
```

```python
# Problem 4 : Plot the count of Survivors and Non-Survivors
```

```python
sns.catplot(x='alive',kind='count',height=6,data=data)
```

```python
# Problem 5: Plot the count of survivors as well as non-survivors on the basis of Pclass
```

```python
sns.catplot(x='alive',hue='pclass',kind='count',height=6,aspect=1.6,data=data)
```

```python
# Problem 6: Plot a scatterplot between age and fare column on the basis of both Pclass and Sex column
```

```python
sns.relplot(y='fare',x='age',hue='sex',style='pclass',size='pclass',height=6,aspect=1.6,sizes=(50,100),data=data,)
```
